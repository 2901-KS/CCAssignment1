```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# QUESTION 1
M = float(input("Enter a value for M: "))
x = np.linspace(-10, 10, 100)
y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, 'r--', label=r'$y = Mx^2$')
plt.plot(x, y2, 'b-', label=r'$y = M \sin(x)$')
plt.title("Mathematical Functions")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.grid()
plt.show()
```

```python
    np.random.seed(102317158)
    data = np.random.randn(50)

    fig, axs = plt.subplots(2, 2, figsize=(10, 8))

    # Line plot of cumulative sum
    axs[0, 0].plot(np.cumsum(data), 'g-', label="Cumulative Sum")
    axs[0, 0].set_title("Cumulative Sum Plot")
    axs[0, 0].set_xlabel("Index")
    axs[0, 0].set_ylabel("Cumulative Sum")
    axs[0, 0].grid()
    axs[0, 0].legend()

    # Scatter plot
    axs[0, 1].scatter(range(50), data, color='r', label="Random Noise")
    axs[0, 1].set_title("Scatter Plot")
    axs[0, 1].set_xlabel("Index")
    axs[0, 1].set_ylabel("Value")
    axs[0, 1].grid()
    axs[0, 1].legend()

    plt.tight_layout()
    plt.show()
```

```python
def sales_data_plots():
    url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
    df = pd.read_csv(url)

    # Line plot of total profit
    plt.figure(figsize=(8, 5))
    plt.plot(df["total_profit"], 'g-', marker='o', label="Total Profit")
    plt.title("Total Profit Over Months")
    plt.xlabel("Month")
    plt.ylabel("Total Profit")
    plt.legend()
    plt.grid()
    plt.show()

    # Multiline plot for all product sales
    plt.figure(figsize=(8, 5))
    for col in df.columns[1:-1]:
        plt.plot(df[col], marker='o', label=col)
    plt.title("Product Sales Over Months")
    plt.xlabel("Month")
    plt.ylabel("Sales")
    plt.legend()
    plt.grid()
    plt.show()

    # Bar chart for all features
    df.set_index("month_number").plot(kind='bar', figsize=(10, 6))
    plt.title("Sales Data Features")
    plt.xlabel("Month")
    plt.ylabel("Values")
    plt.grid(axis='y', linestyle='--')
    plt.show()

sales_data_plots()
```
